# 固有名詞ベクトルのサンプリング

```python
import pickle
import numpy as np
import random 
```

```python
f = open('../koyuu_noun_dic.binaryfile','rb')
noun_dic= pickle.load(f)
```

```python
noun_vec_list = list(noun_dic.values())
```

```python
# 全体の固有名詞の内、半分をランダムに取得

sample_list = random.sample(noun_vec_list, int(len(noun_vec_list)/2))
```

```python
len(sample_list)
```

```python
# データの型を変更

re_noun_vec_list = []

for sample in sample_list:
    a = sample.reshape(50,)
    re_noun_vec_list.append(a)

re_noun_vec_list = np.array(re_noun_vec_list)
```

```python
len(re_noun_vec_list)
```

```python
# ランダムに取得した固有名詞を保存

f = open('../koyuu_noun_sample_dic.binaryfile','wb')
pickle.dump(re_noun_vec_list, f)
f.close
```
